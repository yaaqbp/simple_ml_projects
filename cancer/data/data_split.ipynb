```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('breast-cancer.csv')
```

```python
df.describe()
```

```python
df.head()
```

```python
df.diagnosis.value_counts()
```

```python
df[df.isnull()].sum()
```

```python
df_train_val, df_test = train_test_split(df, test_size = 0.2, stratify = df.diagnosis)
df_train, df_val = train_test_split(df_train_val, test_size = 0.25, stratify = df_train_val.diagnosis)
```

```python
df_train.reset_index(drop = True, inplace = True)
df_val.reset_index(drop = True, inplace = True)
df_test.reset_index(drop = True, inplace = True)
```

```python
df_test
```

```python
df_train.to_csv('data_train.csv', index = False)
df_val.to_csv('data_val.csv', index = False)
df_test.to_csv('data_test.csv', index = False)
```
